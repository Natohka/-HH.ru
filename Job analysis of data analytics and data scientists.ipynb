Возможно есть полностью дублирующие друг друга вакансии, которые отрываются без отправления в архив предыдущей, но такие вакансии в проверку не попали.  Изучив информацию о работе сайта hh.ru можно преположить следующие причины появления таких дубликатов в датафреймах:
1. Вакансия ушла в архив, потому что у работодателя не подключена услуга по её обновлению;
2. Работадатель сам удалил вакансию, найдя нужного кандидата, а затем вернул её (кандидат не подошел, оффер не утроил и др.);
3. Для компании требуется много различных специалистов одного и того же профиля, поэтому закрыв одну вакансию работодатель размещает новую с теми же требованиями, а не обновляет старую (если открывается стартап, новый отдел, либо если компания крупная и множество человек со схожими требованиями работают над одними проектами).  

Удалить дубликаты можно бы в первом описанном выше случае, но по имеющимся у нас данным невозможно определить причину их появления, потому что удаление данных для остальных причин будет ошибкой.

Т.о. столбец `salary_bin` является заполненным только если заполены оба столбца `salary_from` и `salary_to` и число записей с "ЗП не указана" в нем совпадает со количеством отсутсвующих данных для столбца `salary_to`.
Заполнять отсутвующие данные для столбцов `salary_from` и `salary_to` мы не будем, т.к. для каждой компании вилка заработных плат может быть очень широкой и отличатся сильно.  
Создадим новый столбец со средней заработной платой, которую найдем по следующему принципу:
- если оба столбца `salary_from` и `salary_to` пустые, то и новый столбеу оставить пустым;
- если значения есть хотя бы в одном столбце, то средняя зарплата должна быть равна данному значению;
- если оба столбца заполнены, то среднее значение заполняется как сумма этих столбцов / 2.

```python
def calculate_avg_salary(row):
    '''Функция считает среднюю заработную плату на основании двух столбцов salary_from и salary_to'''
    
    # Извлекаем значения минимальной и максимальной заработной платы из строки
    min_salary = row['salary_from']
    max_salary = row['salary_to']
    
    # Если оба значения заработной платы отсутствуют, возвращаем NaN
    if pd.isna(min_salary) and pd.isna(max_salary):
        return np.nan
    # Если отсутствует минимальная заработная плата, возвращаем максимальную
    elif pd.isna(min_salary):
        return max_salary
    # Если отсутствует максимальная заработная плата, возвращаем минимальную
    elif pd.isna(max_salary):
        return min_salary
    # Если оба значения присутствуют, рассчитываем и возвращаем среднюю заработную плату
    else:
        return (min_salary + max_salary) / 2
```

Напишем функцию, которая преобразует строку навыков в список названий навыков в нижнем регистре.

```python
def custom_split(s):
    '''Функция преобразует навыки, убирает лишние символы и переводит навыки в нижний регистр'''
    
    # Убираем квадратные скобки из строки
    s = s.strip("[]")
    
    # Разбиваем строку на элементы, убираем лишние пробелы и одинарные кавычки, переводим в нижний регистр
    items = [item.strip().strip("'").lower() for item in s.split(",") if item != '']
    
    # Возвращаем список преобразованных навыков
    return items
```

Запишем функцию, которая разделяет скилы на "мягкие" и "твердые"

```python
def split_and_classify_skills(skills, soft_skills_set):
    '''Функция пройдет по колонке со навыками, сохранит soft в одну колонку, если их нет, то в hard. 
    Заменит "[]" на 'нет данных' и посчитает количество твердых и мягких навыков, сохранив это в двух колонках'''
    
    hard_skills = []
    soft_skills = []

    # Заменим все значения "[]" на 'нет данных'
    if not skills.strip("[]").strip():
        return 'нет данных', 'нет данных', 0, 0
    
    # Пройдемся по каждому навыку в строке
    for skill in custom_split(skills):
        # Если навык присутствует в наборе soft_skills_set, добавим его в soft_skills
        if skill in soft_skills_set:
            soft_skills.append(skill)
        # В противном случае добавим его в hard_skills
        else:
            hard_skills.append(skill)
    
    # Возвращаем строку с hard_skills или 'нет данных', строку с soft_skills или 'нет данных', 
    # количество hard_skills и количество soft_skills
    return (
        ', '.join(hard_skills).title() if hard_skills else 'нет данных', 
        ', '.join(soft_skills).title() if soft_skills else 'нет данных',
        len(hard_skills), 
        len(soft_skills)
    )
```

- Применим функцию, чтобы создать новые колонки `hard_skills` и `soft_skills`.

```python
data_analytics[['hard_skills', 'soft_skills','num_hard_skills', 'num_soft_skills']] = data_analytics['key_skills'].apply(
    lambda x: pd.Series(split_and_classify_skills(x, soft_skills_set_da))
)
data_scientists[['hard_skills', 'soft_skills','num_hard_skills', 'num_soft_skills']] = data_scientists['key_skills'].apply(
    lambda x: pd.Series(split_and_classify_skills(x, soft_skills_set_ds))
)
```

Для выявления самых расспространненых навыков напишем функцию, которая будет проходится по колонке с навыками и считать их количество.

```python
def process_skills(dataframe: pd.DataFrame, target_column: str = 'soft_skills') -> pd.DataFrame:
    '''Функция получает dataframe, преобразует значения в колонке target_column в список строковых переменных навыков, 
    затем считает частоту одинаковых навыков и выводит df с двумя колонками - навыки и их число'''
    
    all_skills = []
    
    # Проход по каждой строке в dataframe и сбор всех навыков
    for _, row in dataframe.iterrows():
        skills = [skill.strip().lower() for skill in row[target_column].split(',')]
        all_skills.extend(skills)
    
    # Подсчет частоты каждого навыка
    skill_counts = Counter(all_skills)
    
    # Создание нового DataFrame с двумя колонками: навыки и их количество
    return pd.DataFrame(list(skill_counts.items()), columns=["skills", "number"])
```

Преобразуем функцию использованную выше так, чтобы она учитывала грейд пользователя

```python
def process_skills_grade(dataframe: pd.DataFrame, experience: str = None, target_column: str = 'soft_skills') -> pd.DataFrame:
    '''Функция получает dataframe, проверяет колонку experience каждой строки (если experience не введен, то берет все строки)
    и преобразует значения в колонке target_column в список строковых переменных навыков, 
    затем считает частоту одинаковых навыков и выводит df с тремя колонками - experience, навыки и их число'''
    
    all_skills = []
    experiences = []

    # Проход по каждой строке в dataframe и сбор всех навыков с учетом опыта
    for _, row in dataframe.iterrows():
        if experience is None or row['experience'] == experience:
            skills = [skill.strip().lower() for skill in row[target_column].split(',')]
            all_skills.extend(skills)
            experiences.extend([row['experience']] * len(skills))

    # Подсчет частоты каждого навыка с учетом опыта
    skill_counts = Counter(zip(experiences, all_skills))
    
    # Преобразуем Counter в DataFrame
    df_skills = pd.DataFrame(list(skill_counts.items()), columns=["experience_skill", "number"])
    df_skills[['experience', 'skills']] = pd.DataFrame(df_skills['experience_skill'].tolist(), index=df_skills.index)
    df_skills = df_skills.drop(columns=['experience_skill'])

    # Возвращаем DataFrame с колонками experience, skills и number
    return df_skills[['experience', 'skills', 'number']]
```

```python
# Обработка и подсчет частоты твердых навыков с учетом опыта
common_hard_skills_da_grade = process_skills_grade(data_analytics, target_column='hard_skills')

# Фильтрация данных, исключая строки с навыком "нет данных"
common_hard_skills_da_grade = common_hard_skills_da_grade.query('skills != "нет данных"')

# Сортировка данных по количеству навыков в порядке убывания и отображение топ-10 с применением стилизации
common_hard_skills_da_grade.sort_values(by='number', ascending=False).head(10).style.hide_index().background_gradient()
```

Построим таблицы для Аналитиков данных и для Специалистов по Data Science с информацией о количестве требуемых специалистов, учитывая их грейд, и их долей от общего числа вакансий с разбивкой по месяцам размещения вакансии

```python
# Группировка данных по опыту и месяцу, подсчет количества вакансий
month_data_analytics_grade = data_analytics.groupby(['experience','month']).size().reset_index(name='number_of_vacancies')

# Подсчет общего количества вакансий для каждого уровня опыта
total_vacancies = month_data_analytics_grade.groupby('experience')['number_of_vacancies'].transform('sum')

# Вычисление процента вакансий для каждого уровня опыта
month_data_analytics_grade['percentage'] = (month_data_analytics_grade['number_of_vacancies'] / total_vacancies) * 100

# Определение столбцов для исключения из выделения
columns_to_exclude = month_data_analytics_grade.columns.difference(['month', 'experience'])

# Применение градиента и форматирование
month_data_analytics_grade.style.hide_index().background_gradient(
    subset=columns_to_exclude
).format({'percentage': '{:.2f} %'})
```

```python
# Группировка данных по опыту и месяцу, подсчет количества вакансий
month_data_scientists_grade = data_scientists.groupby(['experience', 'month']).size().reset_index(name='number_of_vacancies')

# Подсчет общего количества вакансий для каждого уровня опыта
total_vacancies = month_data_scientists_grade.groupby('experience')['number_of_vacancies'].transform('sum')

# Вычисление процента вакансий для каждого уровня опыта
month_data_scientists_grade['percentage'] = (month_data_scientists_grade['number_of_vacancies'] / total_vacancies) * 100

# Определение столбцов для исключения из выделения
columns_to_exclude = month_data_scientists_grade.columns.difference(['month', 'experience'])

# Применение градиента и форматирование
month_data_scientists_grade.style.hide_index().background_gradient(
    subset=columns_to_exclude
).format({'percentage': '{:.2f} %'})
```

Мы видим, что в феврале в данных по одной вакансии, июнь представлен не полностью. Поэтому проверем как публиковали вакансии на протяжении остальных месяцев и нет ли пропуска дней.

```python
# Группировка данных по опыту и месяцу, подсчет уникальных дат
unique_dates_per_month_da_grade = data_analytics.groupby(['experience', 'month'])['date'].nunique().reset_index()

# Скрытие индекса и применение градиента к столбцу 'date'
unique_dates_per_month_da_grade.style.hide_index().background_gradient(subset=['date'])
```

```python
# Группировка данных по опыту и месяцу, подсчет уникальных дат
unique_dates_per_month_ds_grade = data_scientists.groupby(['experience', 'month'])['date'].nunique().reset_index()

# Скрытие индекса и применение градиента к столбцу 'date'
unique_dates_per_month_ds_grade.style.hide_index().background_gradient(subset=['date'])
```

Объединим таблицы с информацией о количестве требуемых специалистов и их долей от общего числа вакансий с разбивкой по месяцам размещения вакансии с таблицами по количеству дней публикации вакансий.  
После этого посчитаем сколько вакансий публикуется в день в зависимости от месяца. Для этого все опубликованные вакансии за месяц разделим на число активных дней в данном месяце.

```python
# Объединение данных для каждого уровня опыта и месяца
month_data_analytics_grade_merge = pd.merge(unique_dates_per_month_da_grade, month_data_analytics_grade, how='inner')
```

```python
# Вычисление количества вакансий в день
month_data_analytics_grade_merge['vacancies_per_day'] = round(
    month_data_analytics_grade_merge['number_of_vacancies'] / month_data_analytics_grade_merge['date'], 2
)

# Подсчет общего количества вакансий в день для каждого уровня опыта
total_vacancies = month_data_analytics_grade_merge.groupby('experience')['vacancies_per_day'].transform('sum')

# Вычисление процента вакансий в день для каждого уровня опыта
month_data_analytics_grade_merge['percentage_vpd'] = (
    month_data_analytics_grade_merge['vacancies_per_day'] / total_vacancies
) * 100

# Определение столбцов для исключения из выделения
columns_to_exclude = month_data_analytics_grade_merge.columns.difference(['experience', 'month'])

# Применение градиента и форматирование
month_data_analytics_grade_merge.style.hide_index().background_gradient(
    subset=columns_to_exclude
).format({
    'percentage': '{:.2f} %',
    'vacancies_per_day': '{:.2f}',
    'percentage_vpd': '{:.2f} %'
})
```

```python
# Объединение данных для каждого уровня опыта и месяца
month_data_scientists_grade_merge = pd.merge(unique_dates_per_month_ds_grade, month_data_scientists_grade,  how='inner')
```

```python
month_data_scientists_grade_merge['vacancies_per_day'] = round(
    month_data_scientists_grade_merge['number_of_vacancies']/month_data_scientists_grade_merge['date'], 2
    )
total_vacancies = month_data_scientists_grade_merge.groupby('experience')['vacancies_per_day'].transform('sum')
month_data_scientists_grade_merge['percentage_vpd'] = (
    month_data_scientists_grade_merge['vacancies_per_day'] / total_vacancies
) * 100
columns_to_exclude = month_data_scientists_grade_merge.columns.difference(['experience','month'])
month_data_scientists_grade_merge.style.hide_index().background_gradient(
    subset=columns_to_exclude
    ).format({'percentage':'{:.2f} %','vacancies_per_day':'{:.2f}','percentage_vpd':'{:.2f} %'})
```
